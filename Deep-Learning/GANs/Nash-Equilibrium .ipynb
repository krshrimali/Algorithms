## Nash Equilibrium Concept 

Nash Equilibirum is a stage in $>2$ player game where each players know the equilibrium strategy of other player in the game. It's a popular concept in Game Theory.

Reference: https://en.wikipedia.org/wiki/Nash_equilibrium

```python
# necessary imports
import numpy as np
import matplotlib.pyplot as plt
```

```python
# player - 1 (x)
# player - 2 (y)

x = (np.random.random(1) * np.random.randint(-5, 5))[0]
if(x < 0):
    y = (np.random.random(1) * np.random.randint(0, 5))[0]
else:
    y = (np.random.random(1) * np.random.randint(-5, 0))[0]

print("Values: {}, {}".format(x, y))
```

Functions for Player-1: $$p1 = x*y$$ And for Player-2: $$p2 = -x*y$$

We use partial derivatives to update the values.

$$x = x - \frac{\partial p1}{\partial x} * \text{lr}$$

$$y = y + \frac{\partial p2}{\partial y} * \text{lr}$$

Where: $\text{lr}$ = learning rate. 

```python
# functions for player 1 and player 2
p1 = x*y
p2 = -x*y

def update_x(y):
    del_p1_x = y
    return del_p1_x

def update_y(x):
    del_p2_y = -x 
    return del_p2_y
```

```python
lr = 0.01
iterations = 1000 # number of iterations

values_x = []
values_y = []

for i in range(iterations):
    values_x.append(x)
    values_y.append(y)
    
    # update x and y values
    x = x - update_x(y) * lr
    y = y - update_y(x) * lr
    
    if(i % 100 == 0):
        print('Epoch: {}, x: {}, y: {}'.format(i, x, y))
```
