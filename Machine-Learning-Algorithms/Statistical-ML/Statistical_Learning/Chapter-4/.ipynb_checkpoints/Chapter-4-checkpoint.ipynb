```python
# necessary imports
import numpy as np
import matplotlib.pyplot as plt
import scipy
import collections
```

```python
x_axis = np.arange(-3, 3, 0.01)
y_axis = scipy.stats.norm.pdf(x_axis, 0, 1)
```

```python
# just having some fun with try and except
try:
    assert(len(y_axis) == len(x_axis))
except AssertionError:
    print("length of y_axis ({}) != length of x_axis ({}). Recheck your code.".format(len(y_axis), len(x_axis)))
```

```python
help(scipy.stats.norm.pdf)
```

```python
# check if this forms a normal distribution
plt.plot(x_axis, y_axis)
```

**Key Points**:

1. NumPy's `randn` function generated N number of samples from Gaussian Distribution with `mean=0` and `std_dev=1`
2. Gaussian Distributions' params: mean and standard deviation

```python
# let's check if np.random.randn() gives us mean = 0 and std = 1
sample = np.random.randn(1000)
print("Mean: {}, Standard Deviation: {}".format(sample.mean(), sample.std()))
```

```python
# multiply 5 (std_dev), add 50 (mean)
data = 5 * np.random.randn(10000) + 50
print("Mean: {}, Standard Deviation: {}".format(data.mean(), data.std()))
```

```python
# let's plot the histogram and see how the data varies
plt.hist(data, bins=1000) # x-axis (np.unique(data)) and y-axis freq(np.unique(data))
plt.show()
```

**Key Points**:

1. Variance - how much (on avg) the data differs from the mean of the data
2. Variance - measure of spread of the data distribution.
    * Low Variance - values spread out around the mean
    * High Variance - values spread out from the mean

```python
# demonstration of variance
data = np.random.randn(1000)
print("Variance: ", np.var(data))
```

```python
x_axis = np.arange(-3, 3, 0.001)
# low variance
plt.plot(x_axis, scipy.stats.norm.pdf(x_axis, 0, 0.5), label = 'low variance')
# high variance
plt.plot(x_axis, scipy.stats.norm.pdf(x_axis, 0, 1.5), label = 'high variance')
plt.legend()
plt.show()
```
