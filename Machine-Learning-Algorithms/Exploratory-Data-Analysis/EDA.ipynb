**Data Exploration**

Target Variable: `species`
Features: `sepal_length, sepal_width, petal_length, petal_width`

Let's check the correlation among each features. Since this is a classification problem, correlation may not play any role here.

````python
iris.corr()
````

Note: The target variable isn't visible in the matrix because it's a categorical variable. Just in case if, below is the snippet to map values to numerical values.

```python
iris['species'] = iris['species'].map({'setosa' : -1, 'virginica': 0, 'versicolor': 1})
```

````python
iris['species_numerical'] = iris['species'].map({'setosa' : 2, 'virginica': 3, 'versicolor': 4})
````

Let's now check the graphs. Starting with scatter plots.

````python
sns.set_style("whitegrid");
sns.FacetGrid(iris, hue="species", size=8).map(plt.scatter, "sepal_length", "sepal_width").add_legend()
````

````python
sns.set_style("whitegrid");
sns.FacetGrid(iris, hue="species", size=8).map(plt.scatter, "petal_length", "petal_width").add_legend()
````

Clearly, `petal_length` and `petal_width` graph gives more information about the separation between `setosa` and `versicolor`. 

Let's look at a 3D plot

````python
from mpl_toolkits.mplot3d import Axes3D
````

````python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(iris['petal_length'], iris['petal_width'], iris['sepal_length'], c=iris['species_numerical'], s=60)
# ax.view_init(30, 185)
plt.show()
````

Let's analyze Pair Plots.

````python
plt.close();
sns.set_style("darkgrid");
sns.pairplot(iris, hue="species", size=3);
# to drop species_numerical, use: iris.drop(iris['species_numerical'])
plt.show()
````

The diagonal graphs are the PDFs. We can clearly see that `petal_length` and `petal_width` make the best separation between `setosa` and `versicolor` and moderate separation between `versicolor` and `virginica`.
